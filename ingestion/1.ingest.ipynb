```python
import requests
import pandas as pd

url = "https://api.octopus.energy/v1/accounts/A-7C5BE1AB/"
api_key = ""

response = requests.get(url, auth=(api_key, ''))

d = []

if response.status_code == 200:
    output_dict = response.json()

    for property in output_dict.get('properties', []):
        for electricity_meter_point in property.get('electricity_meter_points', []):
            for agreement in electricity_meter_point.get('agreements', []):
                print("tariff_code", agreement['tariff_code'])
                print("valid_from", agreement['valid_from'])
                print("valid_to", agreement['valid_to'])
                print("product_code", agreement['tariff_code'][5:-2])
                d.append(
                    {
                        'tariff_code': agreement['tariff_code'],
                        'valid_from': agreement['valid_from'],
                        'valid_to': agreement['valid_to'],
                        'product_code': agreement['tariff_code'][5:-2],
                        'type': 'electricity_meter_points'
                    }
                )

        gas_meter_points = property.get('gas_meter_points', [])
        if gas_meter_points:
            for gas_meter_point in gas_meter_points:
                for agreement in gas_meter_point.get('agreements', []):
                    print("tariff_code", agreement['tariff_code'])
                    print("valid_from", agreement['valid_from'])
                    print("valid_to", agreement['valid_to'])
                    print("product_code", agreement['tariff_code'][5:-2])
                    d.append(
                        {
                            'tariff_code': agreement['tariff_code'],
                            'valid_from': agreement['valid_from'],
                            'valid_to': agreement['valid_to'],
                            'product_code': agreement['tariff_code'][5:-2],
                            'type': 'gas_meter_points'
                        }
                    )

    print(d)
    df = pd.DataFrame(d)
    df.to_csv('../containers/processed/account.csv')
else:
    print(f"Error: {response.status_code}")
    print("Response:", response.text)
```
